```python
# calculate the IQR
Q1 = df['fare'].quantile(0.25)
print("Q1 is :,", Q1)
Q3 = df['fare'].quantile(0.75)
print("Q3 is :,", Q3)

IQR = Q3 - Q1
print("IQR is :,",IQR)
```

```python
outliers = (df['fare'] < (Q1 - 1.5 * IQR)) | (df['fare'] > (Q3 + 1.5 * IQR))
outliers
```
